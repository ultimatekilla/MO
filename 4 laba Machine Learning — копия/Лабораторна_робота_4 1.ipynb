##__Навчання з вчителем (Supervised learning)__

>__Навчання з вчителем (навчання на розмічених данних)__ - це напрямок машинного навчання, що об'єднує алгоритми і методи побудови моделей на основі набору прикладів, пари «відомий вхід - відомий вихід». Машина навчається на попередньо розмічених данних.

___Загальна постановка навчання з вчителем:___

Для навчальної вибірки $X=(x_i,y_i)^l_{i=1}$ необхідно знайти такий алгоритм $a\in \mathbb A$, на якому буде мінімізуватися функціонал помилки: $$Q(a,X)\longrightarrow \min_{a\in \mathbb A}$$.

<center><img src="https://www.diegocalvo.es/wp-content/uploads/2020/02/supervised-learning.png" width="600"></center>

В залежності від множини можливих результатів $\mathbb Y$, задачі поділяються на декілька типів:

* __Задачі регресії__

$\mathbb Y$ є дійсною змінною
<center><img src="https://higherlogicdownload.s3.amazonaws.com/IMWUC/UploadedImages/92757287-d116-4157-b004-c2a0aba1b048/linear-regression-in-machine-learning.png" width="400"></center>

* __Задача бінарної класифікації__

В такій задачі простір результатів складається з пари $\mathbb Y=\{0,1\}$. Множина об'єктів, що мають один з пари результатів, називається ___класом___. Тобто, необхідно віднести об'єкти (класифікувати) до одного з цих класів.

<center><img src="https://www.kdnuggets.com/wp-content/uploads/toy-binary-dataset.png" width="400"></center>

* __Задача множинної класифікації__

Класів більше ніж два

<center><img src="https://miro.medium.com/max/1506/1*JAXmOAImcf683aXaBDPPVg.jpeg" width="400"></center>

* __Задача ранжування__

Формальна постановка задачі:

Є вибірка, яка складається з $n$ елементів: $$x_1, . . . , x_n.$$ Вона складається з об'єктів, які мають факторний опис, як і в задачах регресії та класифікації. Ключова відмінність ранжування полягає в цільовій змінній: кожному об'єкту необхідно співвіднести пари виду:$$\{(i, j) :x_i< x_j\}.$$ Набір таких пар і є цільовою змінною. В задачі необхідно побудувати модель $a(x)$, таку, що:$$x_i< x_j \Longrightarrow a(x_i)< a(x_j).$$
Таким чином, по виходам необхідно встановити порядок, а величина результатів не має значення.


##__Задачі класифікації__

Задача класифікації в машинному навчанні — це задача співставлення об'єкту до одного з наперед заданих класів, на основі його формалізованих ознак. Кожен з об'єктів в такій задачі представляє собою вектор в  $N$-вимірному просторі, кожен вимір якого є опис однієї ознаки об'єкта.

Для навчання класифікатору необхідно попередньо поділити всю вибірку даних на тренувальну та тестову, де наперед визначені класи.

>__Найпоширеніші алгоритми для задач класифікації__

* $k$-найближчих сусідів (kNN - $k$-Nearest Neighbors)
* Дерево рішень (DT-Decision Tree)
* Логістична регресія (Logistic Regression)
* Метод опорних векторів (SVM - Support Vector Machine)
* Линійний дискримінантний аналіз LDA (Linear Discriminant Analysis).
* Наївний баєсів класифікатор (Naive Bayes Classifier)
* Беггінг та випадковий ліс (Bagging, Random Forest)
* Бустинг та AdaBoost (Boosting)




##__$k$-найближчих сусідів (kNN - $k$-Nearest Neighbors)__

Найпростіший метричний метод в задачах класифікації, його суть полягає в тому, що нова точка відноситься до того ж класу, що і найближча точка з навчальної вибірки.
<center><img src="https://www.machinelearningmastery.ru/img/0-917935-124414.png" width="400"></center>

___Етапи алгоритму:___
1. розрахувати відстань до кожного з об'єктів навчальної вибірки
2. визначити $k$ об'єктів навчальної вибірки, відстань до яких мінімальна
3. визначити клас об'єкта, що класифікується — це клас, який найчастіше зустрічається серед $k$ найближчих сусідів

Для реалізації алгоритму необхідно визначити оптимальну кількість $k$ сусідів та метрику для розрахунку "близькості" елементів.

__МЕТРИКА (відстань)__

* __Евклідова метрика__
$$\rho(x, y)=\sqrt{\sum_{i=1}^n{(x_i-y_i)^2}} $$
* __Манхеттенська метрика ($L_1$)__
$$\rho(x, y)=\sum_{i=1}^n{|x_i-y_i|} $$
<center><img src="https://cdnb.20m.es/mati-una-profesora-muy-particular/files/2013/04/bilateros_2.jpg" width="400"></center>

* __Метрика Мінковського__
$$\rho(x, y)=\left(\sum_{i=1}^n{|x_i-y_i|^q} \right)^{1/q}$$

$q=1$ - Манхетенська метрика

$q=2$ - Евклідова метрика

$q=\infty$ - Метрика Чебишева
<center><img src="https://www.researchgate.net/publication/349155159/figure/fig1/AS:989596292767746@1612949550717/Three-typical-Minkowski-distances-ie-Euclidean-Manhattan-and-Chebyshev-distances.png" width="400"></center>

__Вибір оптимального $k$__

1. В загальному випадку не існує способу визначення найкращого значення для $k$, тому необхідно спробувати кілька значень, перш ніж зупинитися на одному.

2. Невеликі значення $k$ можуть бути зашумленими і піддаватися впливу викидів.

3. Більші значення $k$ матимуть більш плавні межі прийняття рішень, що означатиме меншу дисперсію, але збільшення зміщенності.

4. Використати алгоритм крос-валідації для визначення $k$.

5. Спробувати зберегти значення $k$ непарним, щоб уникнути плутанини між двома класами даних

<center><img src="https://miro.medium.com/max/622/1*fFKbGPgzexQ8Tzpm5FEMrg.png" width="400"></center>

Частота помилок навчання та коефіцієнт помилок валідації-це два параметри, які нам потрібні для доступу до різних $k$-значень. Нижче наведена крива частоти помилок навчання з різним значенням $k$:

<center><img src="https://www.analyticsvidhya.com/wp-content/uploads/2014/10/training-error.png" width="600"></center>

Як бачите, коефіцієнт помилок при $k = 1$ завжди є нульовим для навчальної вибірки. Це тому, що найближчою точкою до будь - якої точки навчальних даних є вона сама. Тому прогноз завжди точний з $k = 1$. Нижче наведена крива помилки валідації зі змінним значенням $k$:
<center><img src="https://www.analyticsvidhya.com/wp-content/uploads/2014/10/training-error_11.png" width="600"></center>

Рівень помилок спочатку зменшується і досягає мінімуму. Після точки мінімуму вона потім збільшується зі збільшенням $k$. Щоб отримати оптимальне значення $k$, можна поділити вибірку на навчальну та тестову, побудувати криву помилки валідації, щоб отримати оптимальне значення $k$.


Алгоритм kNN може потребувати значної кількості  пам'яті для зберігання всих даних. Метод може погано працювати з багатовимірними даними (множиною вхідних змінних), що негативно впливає на ефективність алгоритму, краще його застосовувати лише на найбільш вагомих ознаках. Також неякісні результати роботи алгоритму можна отримати на зашумлених даних.

Для покращення роботи алгоритму можна вводити ваги на сусідів (зважений алгоритм kNN), як функцію від номеру елемента, або як функцію від відстані. Також можна вводити центроїдний класифікатор: спочатку визначаються центроїди класів, а потім об'єкт класифікується в залежності від близькості до певного центроїда.

### __Переваги методу__

1. Простий у реалізації
2. Гнучкий вибір функцій/відстані
3. Добре працює з багатокласовим випадком
4. Добре працює на лінійно нероздільних даних
5. Може мати хороші результати на практиці, маючи достатньо репрезентативних даних

### __Недоліки методу__

1. Значні затрати пам'яті, алгоритм зберігає всі навчальні дані
4. Стадія прогнозування може бути повільною (з великою розмірністю $N$)
5. Чутливий до невідповідних ознак, викидів та масштабу даних

Цільове поле, яке називається `custcat`, має чотири можливі значення, які відповідають чотирьом групам клієнтів:

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/teleCust1000t.csv'
df_knn = pd.read_csv(path)
df_knn.head()
```

```python
#Сформуємо масив ознак без цільової змінної
X_knn= df_knn[['region', 'tenure','age', 'marital', 'address', 'income', 'ed', 'employ','retire', 'gender', 'reside']] .values
X_knn[0:5]
```

```python
#Проведемо стандартизацію даних для кращого представлення даних для алгоритму kNN
X_knn = preprocessing.StandardScaler().fit(X_knn).transform(X_knn.astype(float))
X_knn[0:5]
```

```python
# Поділимо вибірку на навчальну та тестову в пропорції 80/20

from sklearn.model_selection import train_test_split
X_knn_train, X_knn_test, y_knn_train, y_knn_test = train_test_split( X_knn, y_knn, test_size=0.2, random_state=4)
print ('Навчальна вибірка:', X_knn_train.shape,  y_knn_train.shape)
print ('Тестова вибірка:', X_knn_test.shape,  y_knn_test.shape)
```

```python
# Імпортуємо бібліотеку для KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# візьмемо k=3
# p=1 - Мангеттенська метрика (по замовчуванню), р=2 - Евклідова метрика, довільне р - метрика Мінковського
k = 3
neigh = KNeighborsClassifier(n_neighbors = k, p=1).fit(X_knn_train,y_knn_train)
neigh
```

```python
# зробимо прогноз

yhat_knn = neigh.predict(X_knn_test)
yhat_knn[0:5]
```

```python
# оцінимо точність прогнозу

from sklearn import metrics
print("Точність на навчальних даних: ", metrics.accuracy_score(y_knn_train, neigh.predict(X_knn_train)))
print("Точність на тестових даних: ", metrics.accuracy_score(y_knn_test, yhat_knn))
```

###__Вибір $k$__

Застосуємо функцію `GridSearchCV` для автоматичного вибору $k$

```python
# імпортуємо Pipeline для лінійного об'єднання необхідних процесів
from sklearn.pipeline import Pipeline
```

```python
# сформуємо модель, яку хочемо навчити
knn_pipe = Pipeline([('scaler', preprocessing.StandardScaler()), ('knn', KNeighborsClassifier())])
```

```python
# визначимо параметр, який будемо обирати (кількість сусідів k від 1-9)
knn_params = {'knn__n_neighbors': range(1, 10)}
```

```python
# імпортуємо GridSearchCV для визначення k
from sklearn.model_selection import GridSearchCV
```

```python
# задамо параметри для GridSearchCV
knn_grid = GridSearchCV(knn_pipe, knn_params,
cv=5)
# параметр сv задає кількість підмножин для крос-валідації
```

```python
# навчаємо модель
knn_grid.fit(X_knn_train, y_knn_train)
```

```python
# визначаємо найкраще k та точність на навчальних даних
knn_grid.best_params_, knn_grid.best_score_
```

```python
# визначаємо точність на тестових даних
metrics.accuracy_score(y_knn_test, knn_grid.predict(X_knn_test))
```

##__Дерево рішень (DT-Decision Tree)__



<center><img src="https://www.jcchouinard.com/wp-content/uploads/2021/11/image-1.png.webp" width="400"></img></center>

Дерево рішень як алгоритм машинного навчання – об'єднання логічних правил виду "Значення ознаки $a$ меньше за $x$ та "Значення ознаки $b$ меньше за $y$… $\Longrightarrow$ _Клас 1_" в структуру даних "Дерево".

Дерева рішень легко інтерпретуються на відміну від інших моделей, які працюють скоріш як "чорний ящик", в який завантажили дані та отримали відповідь. Також дерево рішень працює як з неперервними, так і з категоріальними даними.


___Етапи алгоритму:___

1. Для кожного вузла:

  1.1. Вибирається фактор з набору даних.

  1.2. Обчислюється значимість цього фактору (приріст інформації) при розбитті даних на класи.

  1.3 Повторюється для кожного фактору.

2. Розподіляються дані за фактором, який є найбільш значущим (найбільший приріст інформації).

3. Повторюється процедура, доки ентропія не буде дорівнювати або буде близькою до $0$.











__Ентропія__ – це міра хаосу або невизначеності.
$$H(X)=-\sum_{i=1}^n{p_i\log_2{p_i}}$$
$p_i=\frac{N_i}{N}$ – це частотна ймовірність належності даних до класу $i$, де $N_i$ - кількість об'єктів у класі $N$ - кількість даних всього, $n$ - кількість класів.

<center><img src="https://miro.medium.com/max/750/1*M15RZMSk8nGEyOnD8haF-A.png" width="400"></img></center>

__Інформаційний приріст (Information Gain)__ - величина для виміру зменшення хаосу в додатковій інформації (ознаках/незалежних змінних) цільової змінної/класу:

$$IG=(Y,X)=H(Y)-H(Y|X)$$

<center><img src="https://miro.medium.com/max/954/0*EfweHd4gB5j6tbsS.png" width="400"></img></center>


__Індекс Джині (Gini impurity)__ - метрика розбиття дерева рішень. Ймовірність того, що випадково обраний елемент із набору даних буде невірно класифікований.
$$Gini=1-\sum_{i=1}^n{(p_i)^2}$$

##__Логістична регресія (LR - Logistic Regression))__

Логістична регресія є частинним випадком лінійного бінарного класифікатора, яка повертає значення функції у межах від $0$ до $1$, що можна інтерпретувати як ймовірність $p_+$ того, що об'єкт $\vec{x_i}$ відноситься до классу $"+"$:


$$\large p_+ = P\left(y_i = 1 \mid \vec{x_i}, \vec{w}\right) $$


Прогнозування не просто відповіді, а саме ймовірності віднесення до класу $"+"$ в багатьох задачах є важливою бізнес-вимогою. Наприклад, в задачах кредитного скорингу, де застосовується  логістична регресія, часто прогнозують ймовірність неповернення кредиту ($p_+$). Клієнті сортують по цій передбачуваній ймовірності (за спаданням), і отримують скоркарту — по суті, рейтинг клієнтів від "ненадійних" до "надійних".

За допомогою ЛР ми хочемо спрогнозувати ймовірність $p_+ \in [0,1]$. Лінійна регресія за допомогою МНК дає можливість побудувати лінійний прогноз: $b(\vec{x}) = \vec{w}^T \vec{x} \in \mathbb{R}$.

Для того,т щоб перетворити отримане значення в ймовірність необхідна функція $f: \mathbb{R} \rightarrow [0,1].$

В моделі логістичної регресії береться функція ___"сігмоїда" (логіт-функція)___: $$\sigma(z) = \frac{1}{1 + \exp^{-z}}$$.


<center><img src="https://static.javatpoint.com/tutorial/machine-learning/images/logistic-regression-in-machine-learning.png" width="400"></center>

Позначимо $P(X)$ ймовірність події $X$. Тоді відношення ймовірностей $OR(X)$ визначається з $\frac{P(X)}{1-P(X)}$ — відношення ймовірностей того, відбудеться подія чи ні. Ймовірність та відношення шансів містять однакову інформацію. Проте, $P(X)$ знаходится в межах від $0$ до $1$, а $OR(X)$ - в межах від $0$ до $\infty$.


Якщо прологарифмувати $OR(X)$, то $\log{OR(X)} \in \mathbb{R}$. його і будемо прогнозувати методом МНК.

Логістична регресія буде робити прогноз $p_+ = P\left(y_i = 1 \mid \vec{x_i}, \vec{w}\right)$.


__Крок 1.__ Обчислити значення $w_{0}+w_{1}x_1 + w_{2}x_2 + ... = \vec{w}^T\vec{x}$. (рівняння $\vec{w}^T\vec{x} = 0$ задає гіперплощину, яка ділить ознаки на 2 класи);


__Крок 2.__ Обчислюємо логарифм відношення шансів: $ \log(OR_{+}) = \vec{w}^T\vec{x}$.


__Крок 3.__ Обчислюємо $p_{+}$:

$$\large p_{+} = \frac{OR_{+}}{1 + OR_{+}} = \frac{\exp^{\vec{w}^T\vec{x}}}{1 + \exp^{\vec{w}^T\vec{x}}} = \frac{1}{1 + \exp^{-\vec{w}^T\vec{x}}} = \sigma(\vec{w}^T\vec{x})$$


В правій частині отримали сигмоїд-функцію.

Логістична регресія прогнозує ймовірність віднесення об'єкту до класу "+" як сигмоїд-перетворення лінійної комбінації вектора вагів моделі та вектору ознак об'єкту.

$$\large p_+(x_i) = P\left(y_i = 1 \mid \vec{x_i}, \vec{w}\right) = \sigma(\vec{w}^T\vec{x_i}). $$

###__Види логістичної регресії__

1. ___Бінарна або біноміальна___
> в такій класифікації цільова змінна має два можливих результати: $1$ та $0$.
2. ___Поліноміальна___
> в такій класифікації цільова змінна може мати 3 та більше можливих невпорядкованих результатів або результат, як категоріальну змінну.
3. ___Порядкова регресія___
> в такій класифікації цільова змінна може мати 3 та більше можливих впорядкованих результатів, які  мають кількісну інтерпретацію.

###__Передумови для логістичної регресії__

1. ЛР передбачає, що мультиколінеарність серед незалежних змінних мінімальна або відсутня.

2. ЛР передбачає, що незалежні змінні лінійно зв'язані з логарифмом шансів.

3. ЛР потребує великого розміру вибірки та правильного прогнозування.

4. ЛР передбачає, що спостереження незалежні.

### __Переваги методу__

1. Модель ЛР не тільки класифікує, а дає ймовірності належності класу.

2. ЛР дає не тільки міру того, на скільки релевантний предиктор (розмір коефіцієнту), але і "напрям" асоціації (додатній або від'ємний). ЛР простіше реалізувати, інтерпретувати та єфективно навчити.

3. ЛР є досить ефективною, коли набір даних має линійно відокремлювані функції.

4. В низьковимірному наборі даних, які мають достатню кількість об'єктів для навчання, ЛР не так схильна до перенавчання.

### __Недоліки методу__

1. ЛР перенавчається на багатовимірних наборах даних.

2. Низька якість для нелінійних функцій

3. Не в змозі працювати зі складними (комплексними) відношеннями між змінними, та змінними, які слабко пов'язані з цільовою змінною

##__Оцінка точності класифікації__

###__Матриця невідповідностей (Confusion Matrix)__
<center><img src="https://2.bp.blogspot.com/-EvSXDotTOwc/XMfeOGZ-CVI/AAAAAAAAEiE/oePFfvhfOQM11dgRn9FkPxlegCXbgOF4QCLcBGAs/s1600/confusionMatrxiUpdated.jpg" width="600"></img></center>

__TP__ - правильно прийнята гіпотеза $H_0$

__TN__ - правильно прийнята гіпотеза $H_1$

__FP__ - неправильно знехтувана гіпотеза $H_0$ (Помилка першого роду, $\alpha$-помилка, рівень значущості)

__FN__ - неправильно прийнята гіпотеза $H_0$ (Помилка другого роду, $\beta$-помилка, рівень значущості, $1-\beta$ - потужність критерію)

__Presision (Точність-Влучність)__ - частка правильно спрогнозованих значень. Прогностична значущість позитивного результату

$$\frac{TP}{TP+FP}$$

__Recall (Повнота)__ - чутливість

$$\frac{TP}{TP+FN}$$

__Sensitivity (Чутливість )__ - частка правильно визначених істино позитивних значень як таких, що мають цей стан

$$\frac{TP}{TP+FN}$$

__Specificity (Специфічність)__ - частка правильно визначених істино негативних значень як таких, що не мають цей стан

$$\frac{TN}{TN+FP}$$

__Accuracy (Точність)__ - частка правильно спрогнозованих значень
$$\frac{TP+TN}{TP+TN+FP+FN}$$

__F1 score__ - міра точності тесту, визначається як середньогармонійне влучності та повноти. $F1\in[0,1]$, близькість до $1$ означає наближення до ідеальної влучності та повноти.

$$F1=2\cdot\frac{presision\cdot recol}{presision + recol}=\frac{2TP}{2TP+FP+FN}$$
F-міри не беруть до уваги істинно негативних значень для оцінювання продуктивності бінарного класифікатора.

__Jaccard index (Міра Жаккара)__ - бінарна міра подібності.

Якщо два набори даних мають одні й ті самі елементи, їхній індекс подібності Жаккара дорівнюватиме 1. І навпаки, якщо у них немає спільних елементів, їхня схожість дорівнюватиме 0.

$$\frac{TP}{TP+FP+FN}$$

##__Метод опорних векторів (SVM - Support Vector Machine)__

__Метод опорних векторів (SVM)__ - основна ідея методу заключається у відображенні векторів простору ознак, які відображають об'єкти класифікації, в простір більш високої розмірності. Це пов'язано з вищою лінійною відокремленістю в таких просторах.

Після переходу в простір більшої розмірності, будується відокремлююча ___гіперплощина (Hyperplane)___. При цьому всі вектори, розташовані по різні боки від цієї гіперплощини будуть відноситися до різних класів.

По обидві сторони основної гіперплощини, паралельно їй та на однаковій відстані від неї будуються дві допоміжні гіперплощини, відтань між якими називають ___зазор (Margin)___.

Задача постає в побудові гіперплощини таким чином, щоб максимізувати зазор, де не повинно бути векторів.

Вважається, що така гіперплощина забезпечує найбільш вдалий поділ на класи та мінімізує середню помилку передбачень.

Вектори, які потраплять на границі зазору (тобто будуть лежати на допоміжних гіперплощинах), називають ___опорними векторами (Support Vectors)___


<center><img src="https://images1.russianblogs.com/22/60/607df9d749c6974c92515f95cb96ca4e.png" width="500"></img></center>


###__Види SVM__

1. ___Лінійно роздільний SVM___
> навчальні дані є лінійно роздільними, лнійний класифікатор можна навчити максимізуючи жорстку границю (hard-margin SVM).
2. ___Лінійний SVM___
> навчальні дані не можуть бути лінійно відокремлені, але можуть бути приблизно лінійно відокремлені, тобто будується лінійний класифікатор шляхом максимізації "м'якого" запасу.

###__Алгоритм SVM:__

Необхідно знайти рівняння гіперплощини $w_1x_1+w_2x_2+...+w_nx_n+w_0=0$ в просторі $\mathbb{R}$.

Загальний вид претворення $F$ об'єкту $x$ в мітку класу $Y$: $F(x) = sign(w^Tx-b)$, де $w = (w_1, w_2, …, w_n), b=-w_0$. Після налаштування вагів алгоритму $w$ та $b$ (навчання), всі об'єкти будуть класифіковані по розташуванню відносно побудованої гіперплощини.

Всередині функції $sign()$ міститься лінійна комбінація ознак об'єкту з вагами алгоритму, оскільки SVM відноситься до лінійних алгоритмів. В SVM ваги $w$ та $b$ налаштовуються таким чином, щоб  об'єкти класів лежали якомога далі від гіперплощини, максимізуючи зазор.

<center><img src="https://hsto.org/r/w1560/webt/n-/5u/gy/n-5ugyyyqejoobj9javfcxaycpk.png" width="400"></img></center>

Для налаштування вагів, мінімізуємо їх норму:

$$\Arrowvert w\Arrowvert \rightarrow min$$
$$(w^Tw)/2 \rightarrow min$$

Зазором (margin) між об'єктом $x$ та границею класів називається величина $M=y(w^Tx-b)$. Алгоритм допускає помилку на об'єкті тоді, коли $M$ від'ємне (коли $y$ та $(w^Tx-b)$ різних знаків). Якщо  $M \in (0, 1)$, то об'єкт потрапляє в середину зазору. Если $M > 1$, то об'єкт $x$ класифікується правильно, та знаходиться на деякій відстані від зазору. Тобто:

$$y(w^Tx-b) \geqslant 1$$

У випадку ___hard-margin SVM___ маємо:
$$ \left\{ \begin{array}{ll} (w^Tw)/2 \rightarrow min & \textrm{}\\ y(w^Tx-b) \geqslant 1 & \textrm{} \end{array} \right. $$

### __Переваги методу__

1. добре працює з простором ознак великої розмірності;
2. добре працює з даними невеликого об'єму;
3. алгоритм максимізує зазор, який дозволяє зменшити кількість помилок класифікації;
4. оскільки алгоритм зведено до задачі квадратичного програмування в опуклій області, то така задача завжди має єдиний розв'язок (розподіляюча гіперплощина з визначеними гіперпараметрами завжди єдина).

### __Недоліки методу__

1. довгий час навчання алгоритму (для великих наборів даних);
2. нестійкість до шумів: викиди в навчальних даних стають опорними об'єктами і впливають на побудову гіперплощини;
3. не описані загальні методи побудови ядер та спрямляючих просторів у випадку лінійної неподільності класів.

### Завдання №2

* Проведіть стандартизацію даних
* застосуйте алгоритм kNN для $k=3$, $k=5$, використовуючи евклідову метрику та манхеттенську метрику
* порівняйте результати точності для цих моделей
* визначіть оптимальне $k$ за допомогою `GridSearchCV`

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score

data = pd.read_csv('heart.csv')

scaler = StandardScaler()
features = data.drop('target', axis=1)
features_scaled = scaler.fit_transform(features)
target = data['target']

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=42)

def train_knn(X_train, X_test, y_train, y_test, n_neighbors, metric):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric=metric)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    return accuracy

results = {}
for k in [3, 5]:
    for metric in ['euclidean', 'manhattan']:
        accuracy = train_knn(X_train, X_test, y_train, y_test, k, metric)
        results[f'k={k}, metric={metric}'] = accuracy
        print(f'Accuracy for k={k}, metric={metric}: {accuracy}')

param_grid = {'n_neighbors': range(1, 20), 'metric': ['euclidean', 'manhattan']}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print('Best parameters:', grid_search.best_params_)
print('Best cross-validation score (accuracy):', grid_search.best_score_)

# Висновки
print("\nВисновки:")
print("Застосування GridSearchCV допомогло ідентифікувати оптимальні параметри для моделі kNN.")
print("Це підвищує точність моделі, оптимізуючи вибір k та метрику.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

data = pd.read_csv('heart.csv')

scaler = StandardScaler()
features = data.drop('target', axis=1)
features_scaled = scaler.fit_transform(features)
target = data['target']

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=42)

def train_decision_tree(criterion):
    tree = DecisionTreeClassifier(criterion=criterion, random_state=42)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    plt.figure(figsize=(12, 8))
    plot_tree(tree, filled=True, feature_names=data.columns[:-1], class_names=['No Disease', 'Disease'])
    plt.title(f'Decision Tree using {criterion} criterion')
    plt.show()
    
    return accuracy

accuracy_entropy = train_decision_tree('entropy')
print(f'Accuracy with entropy criterion: {accuracy_entropy}')

accuracy_gini = train_decision_tree('gini')
print(f'Accuracy with gini criterion: {accuracy_gini}')

param_grid = {'max_depth': range(1, 10), 'max_features': range(1, len(data.columns))}
tree = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(tree, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print('Best parameters:', grid_search.best_params_)
print('Best cross-validation score:', grid_search.best_score_)

# Висновки
print("\nВисновки:")
print("GridSearchCV допоміг ідентифікувати оптимальні параметри для моделі дерева рішень.")
print("Вибір критерію і налаштувань впливає на точність моделі, покращуючи здатність моделі уникати перенавчання.")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

data = pd.read_csv('heart.csv')

data = data.sample(frac=1, random_state=42)

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)
print(f'kNN accuracy: {knn_accuracy}')

tree = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42) 
tree.fit(X_train, y_train)
tree_predictions = tree.predict(X_test)
tree_accuracy = accuracy_score(y_test, tree_predictions)
print(f'Decision Tree accuracy: {tree_accuracy}')

plt.figure(figsize=(8, 4))
methods = ['kNN', 'Decision Tree']
accuracies = [knn_accuracy, tree_accuracy]
plt.bar(methods, accuracies, color=['blue', 'green'])
plt.ylabel('Accuracy')
plt.title('Comparison of Classification Methods')
plt.show()

print("\nВисновки:")
print("Порівняння точності двох класифікаційних методів показує, який метод краще підходить для даного датасету.")
print("Це може допомогти прийняти рішення про вибір моделі для подальшого використання в реальних застосуваннях.")
```
