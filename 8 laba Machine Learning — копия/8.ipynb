```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження даних
data_path = 'basket_analysis.csv'
data = pd.read_csv(data_path)

# Видалення колонки "Unnamed: 0"
if "Unnamed: 0" in data.columns:
    data.drop("Unnamed: 0", axis=1, inplace=True)

# Перевірка на наявність відсутніх значень та їх обробка
if data.isnull().sum().sum() > 0:
    data.fillna(False, inplace=True)

# Перетворення булевих значень в числа для простоти підрахунку
data_numeric = data.astype(int)

# Підрахунок загальної кількості покупок для кожного продукту
product_frequencies = data_numeric.sum().sort_values(ascending=False)

# Візуалізація
plt.figure(figsize=(12, 8))
product_frequencies.plot(kind='bar')
plt.title('Частота покупки кожного продукту')
plt.xlabel('Продукти')
plt.ylabel('Частота покупок')
plt.xticks(rotation=90)
plt.show()

# Гістограма розподілу розмірів транзакцій
transaction_sizes = data_numeric.sum(axis=1)
plt.figure(figsize=(10, 6))
plt.hist(transaction_sizes, bins=range(int(transaction_sizes.max() + 1)), alpha=0.75, color='blue')
plt.title('Розподіл розмірів транзакцій')
plt.xlabel('Кількість товарів у транзакції')
plt.ylabel('Частота')
plt.xticks(range(int(transaction_sizes.max() + 1)))
plt.show()
```
